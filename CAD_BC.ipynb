```
import os
from keras.preprocessing.image import ImageDataGenerator

#Pre-procesamiento de imágenes con Keras
batch_size = 10

train_datagen = ImageDataGenerator(rescale=1./255,
                                   brightness_range=(0.3, 0.9),
                                   horizontal_flip=True)

#train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train1_dir,
                                                    target_size=(1000,1000),
                                                    batch_size=batch_size,
                                                    class_mode="binary")
val_generator = val_datagen.flow_from_directory(validation1_dir,
                                                target_size=(1000,1000),
                                                batch_size=batch_size,
                                                class_mode="binary")
test_generator = test_datagen.flow_from_directory(test1_dir,
                                                  target_size=(1000,1000),
                                                  batch_size=batch_size,
                                                  class_mode="binary")
```

```
#Aqruitectura Modelo Neuronal
from keras import layers
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# DECLARE AQUI el modelo LeNet-5 (llamelo lenet)
CAD_BR = Sequential()
CAD_BR.add(Conv2D(4, (5,5), activation='relu', input_shape=(1024,1024,3))) #inputshape es el tamaño de la entrada nxn y el numero de planos de color
CAD_BR.add(MaxPooling2D(pool_size=(9,9), strides=(2,2)))
CAD_BR.add(Dropout(0.4))

CAD_BR.add(Conv2D(8,(5,5), activation='relu'))
CAD_BR.add(MaxPooling2D(pool_size=(7,7), strides=(2,2)))
CAD_BR.add(Dropout(0.4))

CAD_BR.add(Conv2D(16,(5,5), activation='relu'))
CAD_BR.add(MaxPooling2D(pool_size=(5,5), strides=(2,2)))
CAD_BR.add(Dropout(0.4))

CAD_BR.add(Conv2D(16,(7,7), activation='relu'))
CAD_BR.add(MaxPooling2D(pool_size=(3,3), strides=(2,2)))

CAD_BR.add(Flatten())
CAD_BR.add(Dense(90, activation='relu'))
#CAD_BR.add(Dropout(0.25))
CAD_BR.add(Dense(60, activation='relu'))
CAD_BR.add(Dense(30, activation='relu'))
CAD_BR.add(Dropout(0.2))
CAD_BR.add(Dense(2, activation='sigmoid'))

CAD_BR.summary()
```

```
from keras import optimizers
from keras.callbacks import ModelCheckpoint
CAD_BR.compile(loss='categorical_crossentropy',
               optimizer=optimizers.Adam(lr=1e-2),
               metrics=['acc'])

checkpointer = ModelCheckpoint(filepath='CAD_BR113.h5',
                               monitor='val_acc',
                               verbose=0,
                               save_best_only=True,
                               mode='max',
                               period=1)

h = CAD_BR.fit_generator(train_generator,
                         steps_per_epoch=247//10, 
                         epochs=20,
                         validation_data=val_generator,
                         validation_steps=42//10,
                         callbacks=[checkpointer])
```

```
##Gráfica de entrenamiento modelo InceptionResNetV2 (1)##
import matplotlib.pyplot as plt
import numpy as np

epoch_max = np.argmax(h.history['val_acc'])
plt.figure(num=None, figsize=(4, 3))
plt.plot(h.history['acc'], label='train_acc')
plt.plot(h.history['val_acc'], label='val_acc')
plt.legend(loc='lower right')
plt.plot(epoch_max, h.history['val_acc'][epoch_max],'*')
plt.title('Accuracy')
plt.show()

CAD_BR.load_weights('CAD_BR102.h5') #cambiar conforme al modelo (1)

##Porcentaje de Entrenamiento y Validación del Modelo InceptionResNetV2##
loss, acc = CAD_BR.evaluate_generator(train_generator, steps=52) #cambiar conforme al modelo
print('train acc =', acc)
loss, acc = CAD_BR.evaluate_generator(val_generator, steps=52) #cambiar conforme al modelo
print('val acc =', acc)

##Evaluación del Modelo InceptionResNetV2 con set Test##
test_loss, test_acc = CAD_BR.evaluate_generator(test_generator, steps=24) #cambiar conforme al modelo
print('test acc:', test_acc)
```
